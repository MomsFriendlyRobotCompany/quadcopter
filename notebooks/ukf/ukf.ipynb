# Unscented Kalman Filter Equations

- Unscented transformation is a novel method for calculating the statistics of a random variable that undergoes a nonlinear transformation. 
- Extended Kalman filter (EKF) post, when input samples are propagated through the linearized first-order model instead of the non-linear model directly, the distribution is gaussian and that made it possible for EKF to work with non-linear models.
- P = covariance matrix = $\frac {1}{N} \sum ( (X-x)(X-x))^T$
- The unscented transformation is a novel method for calculating the statistics of a random variable that undergoes a nonlinear transformation. It is founded on the intuition that approximating a Gaussian distribution is easier than approximating an arbitrary nonlinear function.

## Prediction Step

Generate Sigma Points:

$$
\begin{align*}
X^{[0]} &= \mu \\
X^{[i]} &= \mu + \Delta x_i \quad \text{for } i = 1, \ldots, n \\
X^{[i+n]} &= \mu - \Delta x_{i} \quad \text{for } i = 1, \ldots, n \\
\Delta x &= \sqrt{(n+\lambda)P}
\end{align*}
$$

Predict Sigma Points:

$$
X_k^{[i]} = f(X_{k-1}^{[i]}, \Delta t) \quad \text{for } i = 0, \ldots, 2n
$$

Predicted Mean and Covariance:

$$
\begin{align*}
\hat{x}_k^- &= \sum_{i=0}^{2n} w_i^{[m]}X_k^{[i]} \\
P_k^- &= \sum_{i=0}^{2n} w_i^{[c]}(X_k^{[i]} - \hat{x}_k^-)(X_k^{[i]} - \hat{x}_k^-)^T + Q_k
\end{align*}
$$

## Update Step

Generate Sigma Points:

$$
\begin{align*}
X^{[0]} &= \hat{x}_k^- \\
X^{[i]} &= \hat{x}_k^- + (\sqrt{(n+\lambda)P_k^-})_i \quad \text{for } i = 1, \ldots, n \\
X^{[i+n]} &= \hat{x}_k^- - (\sqrt{(n+\lambda)P_k^-})_{i} \quad \text{for } i = 1, \ldots, n
\end{align*}
$$

Predict Measurement Sigma Points:

$$
Z_k^{[i]} = h(X_k^{[i]}) \quad \text{for } i = 0, \ldots, 2n
$$

Predicted Measurement Mean and Covariance:

$$
\begin{align*}
\hat{z}_k^- &= \sum_{i=0}^{2n} w_i^{[m]}Z_k^{[i]} \\
P_{z,k}^- &= \sum_{i=0}^{2n} w_i^{[c]}(Z_k^{[i]} - \hat{z}_k^-)(Z_k^{[i]} - \hat{z}_k^-)^T + R_k
\end{align*}
$$

Compute Cross-Covariance:

$$
\begin{align*}
P_{x,z,k}^- = \sum_{i=0}^{2n} w_i^{[c]}(X_k^{[i]} - \hat{x}_k^-)(Z_k^{[i]} - \hat{z}_k^-)^T
\end{align*}
$$


$$
\begin{align*}
& \text{Kalman Gain:} \\
& K_k = P_{x,z,k}^- (P_{z,k}^-)^{-1}
\end{align*}
$$
$$
\begin{align*}
& \text{Update State Estimate and Covariance:} \\
& \hat{x}_k = \hat{x}_k^- + K_k(z_k - \hat{z}_k^-) \\
& P_k = P_k^- - K_kP_{z,k}^-K_k^T
\end{align*}
$$
